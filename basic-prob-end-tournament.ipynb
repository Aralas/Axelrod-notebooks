This notebook shows how to build a probabilistic ending tournament. The documentation for this can be found here: ...

```python
import axelrod as axl
import random  # To set a seed
import seaborn
%matplotlib inline
```

```python
random.seed(1)
reps = 10000
strategies = [axl.Cooperator(), axl.Defector(),
              axl.TitForTat(), axl.Grudger()]
tournament = axl.ProbEndTournament(strategies, prob_end=.5, repetitions=reps)
results = tournament.play()
plot = axl.Plot(results)
plot.boxplot();
```

```python
eco = axl.Ecosystem(results)
eco.reproduce(100)
plot.stackplot(eco);
```

```python
plot.payoff();
```

```python
plot.lengthplot();
```
