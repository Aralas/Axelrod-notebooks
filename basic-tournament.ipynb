This notebook shows how to build a basic tournament and view the results.

```python
import axelrod as axl
import random
%matplotlib inline
```

```python
strategies = [s() for s in axl.basic_strategies]
strategies.append(axl.Random())
tournament = axl.Tournament(strategies)
results = tournament.play()
```

```python
results.ranked_names
```

Visualising the results:

```python
plot = axl.Plot(results)
plot.boxplot();
```

There are many more strategies as well as outputs that the library can create: [take a look at the documentation for further details](http://axelrod.readthedocs.org/en/latest/index.html).
